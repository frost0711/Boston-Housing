### Simple Regression

```python
x = df['RM'].values.reshape(-1,1)
```

```python
y = df['MEDV'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 40)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x_train, y_train)
Reg_predict=reg.predict(x_test)
Reg_predict_train=reg.predict(x_train)
```

```python
from sklearn import metrics
print("R score_test =",metrics.r2_score(y_test,Reg_predict))
print("R score_train =",metrics.r2_score(y_train,Reg_predict_train))
print("MSE = ", metrics.mean_squared_error(y_test,Reg_predict))
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso()
lasso_reg.fit(x_train, y_train)
```

```python
lasso_reg_test = lasso_reg.score(x_test,y_test)
lasso_reg_train = lasso_reg.score(x_train,y_train)
print('lasso_reg_test =',lasso_reg_test, '\nlasso_reg_train =',lasso_reg_train)
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge()
ridge_reg.fit(x_train, y_train)
```

```python
ridge_reg_test = ridge_reg.score(x_test,y_test)
ridge_reg_train = ridge_reg.score(x_train,y_train)
print('ridge_reg_test =',ridge_reg_test, '\nridge_reg_train =',ridge_reg_train)
```

```python
from sklearn.linear_model import ElasticNet
En_reg = ElasticNet()
En_reg.fit(x_train, y_train)
```

```python
En_reg_test = En_reg.score(x_test,y_test)
En_reg_train = En_reg.score(x_train,y_train)
print('En_reg_test =',En_reg_test, '\nEn_reg_train =',En_reg_train)
```

### Multiple Regression

```python
x = df.drop('MEDV',axis=1)
y = df['MEDV']
print(x.shape)
print(y.shape)
```

```python
from sklearn.preprocessing import StandardScaler
# define min max scaler
scaler = StandardScaler()
# transform data
scaled = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 40)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x_train, y_train)
Reg_predict=reg.predict(x_test)
Reg_predict_train=reg.predict(x_train)
```

```python
from sklearn import metrics
print("R score_test =",metrics.r2_score(y_test,Reg_predict))
print("R score_train =",metrics.r2_score(y_train,Reg_predict_train))
print("MSE = ", metrics.mean_squared_error(y_test,Reg_predict))
```
